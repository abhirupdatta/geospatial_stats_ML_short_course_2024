```python
import numpy as np
import matplotlib.pyplot as plt

path = '/Users/zhanwentao/Documents/Abhi/Conference/IBC2024/data/'

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

def linear(x):
    return x

# Generate input data
x = np.linspace(-10, 10, 500)

# Sigmoid
plt.figure(figsize=(6, 4))
plt.plot(x, sigmoid(x), label='Sigmoid', color='orange')
plt.title('Sigmoid Activation Function', fontsize=16)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.savefig(path + 'act_sigmoid.png')

# ReLU
plt.figure(figsize=(6, 4))
plt.plot(x, relu(x), label='ReLU', color='orange')
plt.title('ReLU Activation Function', fontsize=16)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.savefig(path + 'act_relu.png')

# Tanh
plt.figure(figsize=(6, 4))
plt.plot(x, tanh(x), label='Tanh', color='orange')
plt.title('Tanh Activation Function', fontsize=16)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.savefig(path + 'act_tanh.png')

# Linear
plt.figure(figsize=(6, 4))
plt.plot(x, linear(x), label='Linear', color='orange')
plt.title('Linear Activation Function', fontsize=16)
plt.xlabel('Input', fontsize=12)
plt.ylabel('Output', fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.savefig(path + 'act_linear.png')
```
